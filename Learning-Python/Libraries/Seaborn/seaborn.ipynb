### Seaborn, a data visualization library for Python built on top of Matplotlib. It makes it super simple to create statistical graphs of datasets.

```python
import seaborn as sns
```

```python
sns.set_theme(style='darkgrid')
```

```python
sns.color_palette()
```

```python
sns.load_dataset('tips')
```

```python
tips_data = sns.load_dataset('tips')
```

```python
sns.relplot(tips_data, x='total_bill', y='tip', hue='day', size='size', style='smoker', col='sex', col_wrap=2, kind='scatter')
```

```python
sns.scatterplot(tips_data, x='total_bill', y='tip', hue='sex', size='size', style='smoker')
```

```python
dow_data = sns.load_dataset('dowjones')
dow_data
```

```python
sns.lineplot(dow_data, x='Date', y='Price')
```

```python
sns.displot(tips_data, x='tip', col='sex')
```

```python
sns.histplot(tips_data, x='tip', hue='sex')
```

```python
sns.displot(tips_data, x='tip', y='total_bill', hue='day', col='sex', kind='kde', fill=True)
```

```python
sns.ecdfplot(tips_data, x='tip')
```

```python
sns.displot(tips_data, x='tip', kind='hist')
sns.rugplot(tips_data, x='tip', color='orange')
```

```python
sns.catplot(tips_data, x='tip', y='day', kind='strip', hue='sex')
```

```python
sns.swarmplot(tips_data, x='tip', y='day', hue='sex')
```

```python
sns.boxplot(tips_data, x='tip', y='day')
```

```python
sns.boxenplot(tips_data, x='tip', y='day')
```

```python
sns.violinplot(tips_data, x='tip', y='day')
```

```python
sns.barplot(tips_data, x='day', y='tip', hue='sex')
```

```python
sns.pointplot(tips_data, x='tip', y='day', hue='sex')
```

```python
sns.jointplot(tips_data, x='total_bill', y='tip')
```

```python
sns.pairplot(tips_data)
```
